```julia
using GPFlux
using Flux
using Zygote
using Plots
pyplot()
using Random

Random.seed!(4);
```

```julia
X = reshape(range(-π, π, length=100), 1, 100)
y = dropdims(sin.(X), dims=1)
noisy_y = y .+ 0.1*randn(100)
index = Array(1:100)
randperm!(index)

train_x, test_x = X[:, index[1:20]], X[:, index[21:end]]
train_y, test_y = noisy_y[index[1:20]], noisy_y[index[21:end]];
```

```julia
c_mean = ConstantMean()
isorbf_kernel = IsoGaussKernel([0.0], [0.0])
lnoise = [-1.0]
gp = GaussProcess(c_mean, isorbf_kernel, lnoise)

negloglik(gp, train_x, train_y)
```

```julia
ps = params(gp)
```

```julia
# minimize negative loglik

using Flux.Optimise: update!

optimizer = ADAM(0.001)
loss = []
for i in 1:1500
    ll = negloglik(gp, train_x, train_y)
    push!(loss, ll)
    if i==1 || i%100 == 0
        @info "step=$i, loss=$ll"
    end
    gs = gradient(()->negloglik(gp, train_x, train_y), ps)
    for p in ps
        update!(optimizer, p, gs[p])
    end
end

plot(loss, xaxis=(:log))
```

```julia
ps = params(gp)
```

```julia
pred_y, pred_σ2 = predict(gp, X, train_x, train_y)
pred_σ = sqrt.(pred_σ2);
```

```julia
x = reshape(X, :)
xtrain = reshape(train_x, :)
plot(x, y, color=:red, xaxis=("x", (-π, π)), yaxis=("y", (-1.0, 1.0)), label="sin(x)")
plot!(xtrain, train_y, st=:scatter, color=:black, label="data")
plot!(x, pred_y, ribbon=pred_σ, color=:blue, alpha=0.8, label="GP")
xticks!([-π, 0.0, π], ["-π", "0.0", "π"])
yticks!([-1.0, 0.0, 1.0], ["-1.0", "0.0", "1.0"])
title!("GP regression on sin(x)")
plot!(legendfontsize=10)
```
